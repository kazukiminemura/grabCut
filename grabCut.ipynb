```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
ori = cv2.imread('statue.jpg')
ori = cv2.cvtColor(ori, cv2.COLOR_BGR2RGB)
img = ori.copy()
mask = np.zeros(img.shape[: 2], np.uint8)
plt.imshow(img)
print(img.shape)
```

```python
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)
rect = (50, 1, 250, 399)
```

```python
"""
img: input image
mask: mask image 0, 1, 2, 3
rect: rectangle (x, y, w, h) coordinate for foreground
bgdModel: will be used in the algorithm internally
fgdModel: will be used in the algorithm internally
iterCount: number of interations,
mode: options for drawing rectangle or final touchup strockes
"""
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)
```

```python
mask2 = np.where((mask==2) | (mask==0), 0, 1).astype('uint8')
img_m = img*mask2[:, :, np.newaxis]
```

```python
plt.subplot(121)
plt.imshow(img_m)
plt.title("grabcut")
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(ori)
plt.title("original")
plt.xticks([])
plt.yticks([])
```
